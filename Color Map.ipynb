```python
import numpy as np
from hydrogen_calculations import complex_array_to_rgb
import matplotlib.pyplot as plt
```

```python
x, y = np.mgrid[-1:1:100j, -1:1:100j]
r = np.sqrt(x**2 + y**2)
cos_phi = x / r
sin_phi = y / r
psi = 2*r * (cos_phi + 1j * sin_phi)
rgb = complex_array_to_rgb(psi, mag_maps_to='s')
```

```python
plt.imshow(rgb, origin='lower')
```

```python
theta, phi = np.mgrid[0:np.pi:100j, 0:2*np.pi:100j]
psi = np.sin(2*theta) * np.exp(1j*phi)
rgb = complex_array_to_rgb(psi, mag_maps_to='s')
plt.imshow(rgb, origin='upper', extent=[0, 2, 1, 0], aspect=2)
```
