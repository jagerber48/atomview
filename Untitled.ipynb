```python
import numpy as np
from hydrogen_calculations import complex_array_to_rgb
import matplotlib.pyplot as plt
```

```python
x, y = np.mgrid[-1:1:100j, -1:1:100j]
r = np.sqrt(x**2 + y**2)
cos_phi = x / r
sin_phi = y / r
psi = 2*r * (cos_phi + 1j * sin_phi)
rgb = complex_array_to_rgb(psi, mag_maps_to='s', return_alpha=True)
```

```python
plt.imshow(np.transpose(rgb, axes=[1, 0, 2]), origin='lower')
```
